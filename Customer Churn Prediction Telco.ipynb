```python
# Getting towards machine learning part
# We have df1 to be fed to a machine learning model
# But first of all we have to scale columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']
scaler = StandardScaler()
df1[cols_to_scale] = scaler.fit_transform(df1[cols_to_scale])
```

```python
print_unique(df1)
```

```python
# Now we have standardized data
```

```python
# Splitting the data into train and test sets
```

```python
X = df1.drop("Churn", axis = 1)
Y = df1["Churn"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.2, random_state=42)
```

```python
print("Train size X: ",x_train.shape)
print("Train size Y: ",y_train.shape)
print("Test size X: ",x_test.shape)
print("Test size Y: ",y_test.shape)
```

```python
# Checking Logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression()
logr.fit(x_train, y_train)
y_p = logr.predict(x_test)
```

```python
logr.score(x_test, y_test)
```

```python
# Now Checking with Random Forest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
y_p = rfc.predict(x_test)
rfc.score(x_test, y_test)
```

```python
# It has a low accuracy than logistic regression
```

```python
# Now moving towards artificial neural network
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'), # 26 because of 26 columns
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(5, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100)
```

```python
model.evaluate(x_test, y_test)
```

```python
y_p = model.predict(x_test)
y_p
```

```python
y_predict = [1 if p>=0.5 else 0 for p in y_p]
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test, y_predict))
```

```python
cm = confusion_matrix(y_true=y_test, y_pred = y_predict)
```

```python
sns.heatmap(cm, annot = True, fmt = 'd')
```

```python
# Hence, we achieved a 77% acccuracy
```
